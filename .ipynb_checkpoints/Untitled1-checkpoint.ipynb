```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import csv, requests

DECIMAL = 1
stocks = ["IBULHSGFIN", "SBIN", "DLF", "TATASTEEL", "JINDALSTEL", "BANDHANBNK", "PFC", "APOLLOTYRE", "CADILAHC", "M%26MFIN", "TATAMOTORS"]
percent_from_strike = 0.15
expiry = "2020-09-24" #format "yyyy-mm-dd"
```

```python
url = "https://www.moneycontrol.com/stocks/fno/view_option_chain.php?sc_id=ACC&sel_exp_date=2020-09-24"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
curr_price = float(soup.findAll("p", {"class": "r_28 FL"})[0].find("strong").text.strip())
lower = (1-percent_from_strike)*curr_price
upper = (1+percent_from_strike)*curr_price

data = []
columns = []
table = soup.find('table', {"class": "tblopt"})
rows = table.findAll('tr')
cols = rows[1].findAll('th')
cols = [ele.text.strip() for ele in cols]
columns = [ele for ele in cols if ele]
columns.insert(5, 'Strike Price')
table = soup.findAll('table', {"class": "tblopt"})[1]
rows = table.findAll('tr')
for row in rows:
    cols = row.findAll('td')
    cols = [ele.text.strip().replace(',', '') for ele in cols if ele.text.strip()]
    data.append([float(ele) if ele != '-' else 0 for ele in cols]) # Get rid of empty values
df = pd.DataFrame(data, columns=columns)
df
```

```python
relevant = df[df['Strike Price'].between(lower, upper)]
put_relevant = df[df['Strike Price'].between(lower, curr_price)]
call_relevant = df[df['Strike Price'].between(curr_price, upper)]
put_relevant = put_relevant.iloc[:, [5,6]].round(DECIMAL)
call_relevant = call_relevant.iloc[:, [5,0]].round(DECIMAL)
```

```python
url = "https://www.moneycontrol.com/stocks/fno/view_option_chain.php?classic=true"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
options = soup.find(id="stock_code").findAll("option")
count = 0
stocks = []
codes = []
for option in options[1:]:
    stock = option.text.strip()
    code = option['value']
    stocks.append(stock)
    codes.append(code)
df = pd.DataFrame(
{
    'Stocks': stocks,
    'MC Codes': codes
})
```

```python
df
```

```python
df.to_csv('links.csv')
```
